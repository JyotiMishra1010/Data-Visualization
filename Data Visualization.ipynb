```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load the iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# Mapping species to their names
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
iris_df['species'] = iris_df['species'].map(species_map)

# Function to create scatter plots for each pair of features
def scatter_plot(df, feature1, feature2, hue='species'):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x=feature1, y=feature2, hue=hue, style=hue, palette='deep')
    plt.title(f'Scatter Plot: {feature1} vs {feature2}')
    plt.show()

# Function to create histograms for each feature
def histogram(df, feature, hue='species'):
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df, x=feature, hue=hue, multiple="stack", palette='deep')
    plt.title(f'Histogram of {feature}')
    plt.show()

# List of features
features = iris.feature_names

# Create scatter plots for each pair of features
for i in range(len(features)):
    for j in range(i + 1, len(features)):
        scatter_plot(iris_df, features[i], features[j])

# Create histograms for each feature
for feature in features:
    histogram(iris_df, feature)
```
